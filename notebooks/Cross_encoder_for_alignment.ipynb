# Real Data

```python
pairs = []

for chunk in range(len(manifesto_data)):
    for m in sent_tokenize(manifesto_data[chunk]['page_content']):
        for s in sent_tokenize(speech_data[chunk]['page_content']):
            pairs.append((m, s))
```

```python
scores = model.predict(pairs)          # shape: (n_samples, n_labels)
label_mapping = ['contradiction', 'entailment', 'neutral']

# index of the highest-scoring label per pair
best_idx = scores.argmax(axis=1)

# corresponding label and score
best_labels = [label_mapping[i] for i in best_idx]
best_scores = scores[np.arange(len(scores)), best_idx]
```

```python
premis = [p[0] for p in pairs]
hypothesis = [p[1] for p in pairs]
```

```python
df = pd.DataFrame({'premis (m)': premis,'hypothesis (s)':hypothesis,'best_labels': best_labels, 'best_scores': best_scores})
```
